```python
import pandas as pd
```

```python
df=pd.read_csv('glassdoor.csv')
#pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
#Parsing min ,max,average salary
salary=df['Salary Estimate'].apply(lambda a: a.split('(')[0])
min_K = salary.apply(lambda a: a.replace('K','').replace('$',''))
df['min_salary']=min_K.apply(lambda a: int(a.split('-')[0]))
min_hour = min_K.apply(lambda a: a.lower().replace('per hour',''))
df['max_salary']= min_hour.apply(lambda a: int(a.split('-')[1]))
df['avg_salary']=(df.min_salary+df.max_salary)/2
df['avg_salary']=(df.min_salary+df.max_salary)/2
```

```python
#Company_name
df['company_name'] = df.apply(lambda x: x['Company Name'] if x['Rating'] <0 else x['Company Name'][:-4], axis = 1)
```

```python
#State_name
df['JobState']=df['Location'].apply(lambda a: a.split(',')[1])
df['JobState'].value_counts()
df['Same_state']=df.apply(lambda a: 1 if a.Location==a.Headquarters else 0,axis=1)
```

```python
#Company_age
df['Age']=df.Founded.apply(lambda a: a if a<1 else 2021-a)
```

```python
#Python
df['python_n'] = df['Job Description'].apply(lambda a: 1 if 'python' in a.lower() else 0)
df.python_n.value_counts()
```

```python
#R
df['R_studio'] = df['Job Description'].apply(lambda a: 1 if 'r studio' in a.lower() else 0)
df.R_studio.value_counts()
```

```python
#spark 
df['spark_n'] = df['Job Description'].apply(lambda a: 1 if 'spark' in a.lower() else 0)
df.spark_n.value_counts()



```

```python
#aws 
df['aws'] = df['Job Description'].apply(lambda a: 1 if 'aws' in a.lower() else 0)
df.aws.value_counts()

```

```python
#excel
df['excel_n'] = df['Job Description'].apply(lambda a: 1 if 'excel' in a.lower() else 0)
df.excel_n.value_counts()
```

```python
#SQL
df['SQL_n'] = df['Job Description'].apply(lambda x: 1 if 'sql' in x.lower() else 0)
df.SQL_n.value_counts()
```

```python
#Tableau
df['tableau'] = df['Job Description'].apply(lambda x: 1 if 'tableau' in x.lower() else 0)
df.tableau.value_counts()
```

```python
#Removing unnamed column
df_out = df.drop(['Unnamed: 0'], axis =1)
df_out.to_csv('ds_salary_data_cleaned.csv',index = False)
pd.read_csv('ds_salary_data_cleaned.csv')
```
